```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Get the stock prices data for Apple Inc.
stock = yf.Ticker("AAPL")
df = stock.history(period="max")

# Smooth the Close price with a rolling mean of 10 days
df["Close_smoothed"] = df["Close"].rolling(window=10).mean()

# Plot the original and smoothed data
plt.plot(df["Close"], label="Original")
plt.plot(df["Close_smoothed"], label="Smoothed")
plt.legend()
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Get the stock prices data for Apple Inc.
stock = yf.Ticker("AAPL")
df = stock.history(period="max")

# Smooth the Close price with an exponential moving average
df["Close_smoothed"] = df["Close"].ewm(span=10, adjust=False).mean()

# Plot the original and smoothed data
plt.plot(df["Close"], label="Original")
plt.plot(df["Close_smoothed"], label="Smoothed")
plt.legend()
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import TruncatedSVD
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Perform SVD to reduce the dimensionality to 2 dimensions
svd = TruncatedSVD(n_components=2, random_state=42)
X_2d = svd.fit_transform(X)

# Plot the reduced dataset
plt.scatter(X_2d[:, 0], X_2d[:, 1], c=y)
plt.xlabel('SVD1')
plt.ylabel('SVD2')
plt.title('Iris dataset reduced to 2 dimensions with SVD')
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Perform LDA to reduce the dimensionality to 2 dimensions
lda = LinearDiscriminantAnalysis(n_components=2)
X_2d = lda.fit_transform(X, y)

# Plot the reduced dataset
plt.scatter(X_2d[:, 0], X_2d[:, 1], c=y)
plt.xlabel('LDA1')
plt.ylabel('LDA2')
plt.title('Iris dataset reduced to 2 dimensions with LDA')
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Perform PCA to reduce the dimensionality to 2 dimensions
pca = PCA(n_components=2)
X_2d = pca.fit_transform(X)

# Plot the reduced dataset
plt.scatter(X_2d[:, 0], X_2d[:, 1], c=y)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Iris dataset reduced to 2 dimensions with PCA')
plt.show()
```
